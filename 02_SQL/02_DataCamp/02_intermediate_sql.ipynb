## Multiple criteria

### OR

At least ONE condition :

    SELECT *
    FROM coats
    WHERE color = 'green' OR lenght = 'short'

### AND

all conditions :

    SELECT *
    FROM coats
    WHERE color = 'green' AND lenght = 'short'

### OR and AND 

    SELECT *
    FROM coats
    WHERE (color = 'green' OR color = 'red') 
        AND (lenght = 'short' OR lenght = 'long');

### BETWEEN

It's **INCLUSIVE !** 

    SELECT *
    FROM film
    WHERE release_year 
        BETWEEN 1995 AND 2012;

1995 and 2012 are IN !

### BETWEEN + OR or AND

    SELECT *
    FROM film
    WHERE release_year 
        BETWEEN 1995 AND 2012 AND country = 'UK';


### LIKE 

**%**  : match zero, one, or many characters in the text

    SELECT name
    FROM people
    WHERE name LIKE 'Ev%';

   ==> Eva Dupuis, Eve Dupont


**_**  : match single characters

    SELECT name
    FROM people
    WHERE name LIKE 'Ev_';

    ==> Eva, Eve

### NOT LIKE 

dont match 


### wildcard position

'%r'  ==>   Peter
'_a%'  ==>   Tati

### IN 

or operator in list

    SELECT *
    FROM films
    WHERE title IN ('Matrix','Thor')


### NULL values

* IS NULL  : match all missing

* IS NOT NULL : match all no missing

## HAVING vs WHERE
WHERE filters individual records while HAVING filters grouped records


    SELECT country, AVG(budget) AS average_budget
    FROM films
    GROUP BY country
    HAVING AVG(budget) > 1000000000 
    ORDER BY average_budget DESC;
