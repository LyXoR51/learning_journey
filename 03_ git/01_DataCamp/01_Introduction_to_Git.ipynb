# 2. Version history

## Commit structure


* **Commit :** contains metadata such as the author, log message, and commit time
* **tree :** tracks the names and locations of files and directories. like a dictionary, with keys represented as unique identifiers, mapping to files or directories
* **blob Binary Large OBject:** contain data of any kind. Blobs contain a compressed snapshot of the file's contents when the commit happened. 
* **hash :**  unique identifier, 40-character string of numbers. 
    If two files are the same, their hashes will be the sam


## Log

`git log` \
`git log -3`  last 3 commits \
`git log README.md`   specific file \
`git log -2 README.md`   last two for specific file \
`git log --since='apr 2 2025'`  month day year OR `git log --since='04-02-2025'` \
`git log --since='apr 2 2025' --until='apr 5 2025'`  month day year

`git diff report.md`

`git diff --staged report.md`

### between two logs
`git diff hash1 hash2`  Most recent need to be hash2 \
`git diff HEAD~1 HEAD`  HEAD = Last one      HEAD~1  previous

## Restoring and reverting files

### git revert: 
This command undoes changes by creating a new commit that reinstates previous versions of files. 

`git revert HEAD` undo the last commit, need to explain and commit 

`git revert -- no-edit HEAD ` to revert the last commit without opening a text editor 

`git revert -n HEAD` undo and NO COMMIT, so can modofy the file in stag area !

### git checkout: 
to revert a single file to a previous state 

`git checkout HEAD~1 -- report.md`

### git restore:
unstages files, moving them back to the working directory 

`git restore --staged readme.md`  one file 

`git restore --staged .` all files
