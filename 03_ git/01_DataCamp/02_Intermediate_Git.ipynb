# 1. Introduction to branches

## Listing all branches

`git branch`

## Switching between branches

`git switch main`


## Creating a new branch

`git branch speed-test` + `git switch speed-test `

or

`git switch -c speed-test`


## Comparing Branches

`git diff documentation front-end`

## Renaming Branches

`git branch -m txt lw1959-text-bug`     nane to new_name

## Deleting Branches

`git branch -d llm-upgrade`  branch already merged or give error

`git branch -D llm-upgrade`  force delete

## Merging branches

* **source :** the branch we want to merge from
* **destination :** the branch we want to merge into

Move to the destination branch:
`git switch destination`

then merge the source branch :
`git merge source`

or :
`git merge source destination`

## Introduction to remotes

### Cloning a repo

Cloning = make a copy

`git clone path-to-project-repo`

or cloning a local repo 

`git clone /home/user/repo`

Rename the new repo :

`git clone /home/user/repo new_name`


### Cloning a remote

`git clone URL`

`git remote`   name of the remote

`git remote -v`  more information

When cloning a remote, name = origin

### Creating a remote

`git remote add name URL`


### Fetching from a remote

`git fetch origin`   fetch from origin remote. Inlcude all branches


### Fetching a remote branch

`git fetch origin main`   fetch the main branch from origin


### Synchronizing content

after fetching, a synchronizing is necessary with the local repo

`git merge origin`  merge the 'main' branch, by default

### Pulling from a remote

fetch + merge

`git pull origin`

### Pulling a remote branch

`git pull origin dev`


### Pushing to remotes

Push into remote from local_branch
`git push remote local_branch`


sample : `git push origin main`

create a new branch on remote : `git push origin hotfix`
