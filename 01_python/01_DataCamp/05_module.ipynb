# 1. os
---

The `os` module lets us interact with our operating system. We can check the current working directory, list available files, access environment variables, and more

```python
getcwd()	    Return the current working directory as str.
chdir()	        Change the current working directory.
```


## samples : 

```python
# generate work_dir variable
work_dir = os.getcwd()

# generate work_dir variable
os.chdir("path/to/go")
```
